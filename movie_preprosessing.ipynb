```python
# importing essential libraries
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(2)
```

```python
credits.head()
```

```python
# the column cast contains the name of the actors who worked in the film.
credits.head(2)['cast'].values
```

```python
# the column cast contains the name of the crew who worked in the film.(eg- director,producer, etc)
credits.head(2)['crew'].values
```

```python
# Now we will merge these tow data set, as they both contain the information about the movies
movies = movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
movies.info()
```

```python
## Selecting the feature which will be usefull
## Following features will be used
    #genres
    #id
    #keywords
    #title
    #cast
    #crew 
    #overview
movies = movies[['genres','id','keywords','title','overview','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
# in the overview column we have 3 rows as null, so we will just drop them.
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
# Now we will check for any duplicate values
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
## we only need the genres name such as "Action", "Adventure"
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')

# using the ast for changing the format
```

```python
# Creating a function for the same
def convert(obj):
    list_name=[]
    for i in ast.literal_eval(obj):
        list_name.append(i['name'])
    return list_name
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# As we can see from the above that we have changes the content of the feature of genres as we wanted
# Same we will do with the column keyword

movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].cast
```

```python
## We will also do the same procedure on cast column but we will only used the top 3 names i.e we will use first 3 dictionary.
```

```python
def convert3(obj):
    List_ = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            List_.append(i['name'])
            counter+=1
        else:
            break
    return List_
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
## Now changing the crew feature
## we need the name of only those whose job is director
```

```python
def director(obj):
    List_ = []
    
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            List_.append(i['name'])
            break
    return List_
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies.head()
```

```python
movies['overview']
```

```python
# Converting the overview into a list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# removing spaces from the data, to concatenate the first name and last name, so that the model may not get confused.

movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# now our goal is to concatenate all the required feature and convert back into a dataframe which would acts as a tag

movies['tag'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['id','title','tag']]
```

```python
new_df
```

```python
# the tag is in string, converting it into a list
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tag'][1]
```

```python
# converting into lower case
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df.to_csv('movies_final1.csv', index=False)
```
